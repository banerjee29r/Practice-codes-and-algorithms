```python
chitfund['Annualized_Return_of_bid_winner_this_month']=chitfund.Net_amount_recd_by_Bid_winner+sum(chitfund.Amount_returned_to_everyone_in_the_group)
```

```python
chitfund['Annualized_benefit_of_bid_winner_this_month']= chitfund.Annualized_Return_of_bid_winner_this_month - sum(chitfund.Contribution)
```

```python
total_contribution = sum(chitfund.Contribution)
```

```python
chitfund['Annualized_percentage_return_of_bid_winner_this_month']=(chitfund.Annualized_Return_of_bid_winner_this_month/total_contribution)
chitfund
```

```python
chitfund['Annualized_percentage_return_of_bid_winner_this_month'] = pd.Series(["{0:.2f}%".format(val * 100) for val in chitfund['Annualized_percentage_return_of_bid_winner_this_month']], index = chitfund.index)
```
